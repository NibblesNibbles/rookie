```python
#Answer to Question4
y1 = np.array([0.0,0,0,0])
for i in range(0,4):
    y1[i] = 3 + np.random.uniform()
y2 = np.array([0.0,0,0,0])
for i in range(0,4):
    y2[i] = 3 + np.random.uniform()
x = pd.date_range(start = '20190101',end = '20191231',periods = 4)
plt.plot(x,y1,'go-',x,y2,'b^-')
plt.legend(('Apple','Google'),loc = 0)
plt.xlabel('Time')
plt.ylabel('Quarterly Turnover (ten billion)')
plt.title('Quarterly Turnover Report for Apple and Goole in 2019')
plt.show()
```

```python
#Answer to Question5a
df = pd.read_csv('iris.csv')
df1 = df[0:50]
df2 = df[50:100]
df3 = df[100:150]
x11 = df1.SepalLengthCm.to_numpy()
x21 = df1.SepalWidthCm.to_numpy()
x31 = df1.PetalLengthCm.to_numpy()
x41 = df1.PetalWidthCm.to_numpy()
x12 = df2.SepalLengthCm.to_numpy()
x22 = df2.SepalWidthCm.to_numpy()
x32 = df2.PetalLengthCm.to_numpy()
x42 = df2.PetalWidthCm.to_numpy()
x13 = df3.SepalLengthCm.to_numpy()
x23 = df3.SepalWidthCm.to_numpy()
x33 = df3.PetalLengthCm.to_numpy()
x43 = df3.PetalWidthCm.to_numpy()

plt.figure()
ax1 = plt.subplot2grid((40, 30), (0, 0),colspan= 7, rowspan= 5)
ax1.set_title('SepalLength of Iris-setosa',fontsize = 8)
ax1.hist(x11,30)
ax1 = plt.subplot2grid((40, 30), (10, 0),colspan= 7, rowspan= 5)
ax1.set_title('SepalWidth of Iris-setosa',fontsize = 8)
ax1.hist(x21,30)
ax1 = plt.subplot2grid((40, 30), (20, 0),colspan= 7, rowspan= 5)
ax1.set_title('PetalLength of Iris-setosa',fontsize = 8)
ax1.hist(x31,30)
ax1 = plt.subplot2grid((40, 30), (30, 0),colspan= 7, rowspan= 5)
ax1.set_title('PetalWidth of Iris-setosa',fontsize = 8)
ax1.hist(x41,30)
ax1 = plt.subplot2grid((40, 30), (0, 10),colspan= 7, rowspan= 5)
ax1.set_title('SepalLength of Iris-versicolor',fontsize = 8)
ax1.hist(x12,30)
ax1 = plt.subplot2grid((40, 30), (10, 10),colspan= 7, rowspan= 5)
ax1.set_title('SepalWidth of Iris-versicolor',fontsize = 8)
ax1.hist(x22,30)
ax1 = plt.subplot2grid((40, 30), (20, 10),colspan= 7, rowspan= 5)
ax1.set_title('PetalLength of Iris-versicolor',fontsize = 8)
ax1.hist(x32,30)
ax1 = plt.subplot2grid((40, 30), (30, 10),colspan= 7, rowspan= 5)
ax1.set_title('PetalWidth of Iris-versicolor',fontsize = 8)
ax1.hist(x42,30)
ax1 = plt.subplot2grid((40, 30), (0, 20),colspan= 7, rowspan= 5)
ax1.set_title('SepalLength of Iris-virginica',fontsize = 8)
ax1.hist(x13,30)
ax1 = plt.subplot2grid((40, 30), (10, 20),colspan= 7, rowspan= 5)
ax1.set_title('SepalWidth of Iris-virginica',fontsize = 8)
ax1.hist(x23,30)
ax1 = plt.subplot2grid((40, 30), (20, 20),colspan= 7, rowspan= 5)
ax1.set_title('PetalLength of Iris-virginica',fontsize = 8)
ax1.hist(x33,30)
ax1 = plt.subplot2grid((40, 30), (30, 20),colspan= 7, rowspan= 5)
ax1.set_title('PetalWidth of Iris-virginica',fontsize = 8)
ax1.hist(x43,30)
plt.show()
```
