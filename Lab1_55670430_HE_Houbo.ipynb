```python
import matplotlib.pyplot as plt
import numpy as np

#Q1 60.9726171

#Q2
def temperature (t):
    T = 45 + 25*np.e**(-0.056*t)
    return T
#a)
T0_8 = []
for i in range(9):
    T0_8.append(temperature(i))
print(T0_8)
#b)
t = np.arange(0.0,8.2,0.2)
plt.plot(t,temperature(t))
plt.xlabel('time')
plt.ylabel('temperature')
plt.grid(True)
plt.show()
```

```python
#Q3 a)
def predict (time,step):  
    T = np.zeros(len(time))
    for i in range(len(time)):
        if i == 0: T[0] = 70
        else: T[i] = T[i-1] - step*(T[i-1]-45)*0.056
    return T
t= np.arange(0.0,8.5,0.5)
prediction = predict(t,0.5)[16]
print(prediction)

#b)
Error = prediction - temperature(8) 
print('Error = ',Error)

#Q4 a)
plt.plot(t,temperature(t),'go-',t,predict(t,0.50),'b^-')
plt.xlabel('time')
plt.ylabel('temperature')
plt.legend(('True','Predicted'),loc = 0)
plt.grid(True)
plt.show()
#b) 
MSE = np.mean((predict(t,0.5)-temperature(t))**2)
print('MSE = ', MSE)
#c) In the real situation, the time is continus. While in our prediction, the time is discrete with step Δt= 0.5. 
```

```python
#Q5 a)
t = np.arange(0.0,8.73,0.73)
plt.plot(t,temperature(t),'go-',t,predict(t,0.73),'b^-')
plt.xlabel('time')
plt.ylabel('temperature')
plt.legend(('True','Predicted'),loc = 0)
plt.grid(True)
plt.show()
#b) 
MSE = np.mean((predict(t,0.73)-temperature(t))**2)
print('MSE = ', MSE)
#c) The error is bigger compared to the case of Δt= 0.5.
#d) Because the step Δt= 0.73 is larger than that of Δt= 0.5. The larger the step is, the bigger the error is compared to the real situation(continuous time)
```

```python
#Q6 a)
def MSEf (step): 
    t =  np.arange(0.0,8 + step,step)
    mse = np.mean((predict(t,step)-temperature(t))**2)
    return mse 
step = np.arange(0.01,1.01,0.01)
MSElist = np.zeros(len(step))
for i in range(len(step)):
    MSElist[i] = MSEf(step[i])
plt.plot(step,MSElist)
plt.ylabel('MSE')
plt.xlabel('Δt')
plt.grid(True)
plt.show()
#b) Generally, the mean square error increases as Δt increases, with a few exceptions due to the short interval(0.0,8.0).
```
